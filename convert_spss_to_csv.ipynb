```python
import pandas as pd
import pyreadstat
```

```python
path = r"C:\Users\HP\Downloads\Internship\CDHS_2021-2022\KHSR82DT_siblings_recode\KHSR82FL.DTA"
df, meta = pyreadstat.read_dta(path)
```

```python
df_numeric = df.copy()
df_text = df.copy()
if hasattr(meta, 'variable_value_labels') and meta.variable_value_labels:
    for column_name, value_labels in meta.variable_value_labels.items():
        if column_name in df_text.columns:
            df_text[column_name] = df_text[column_name].map(value_labels)
```

```python
df_numeric.to_csv('siblings_numeric.csv', index=False)
df_text.to_csv('siblings_text.csv', index=False)
```
