```python
import re

var_labels = {}

with open("KHNR82FL.MAP", "r", encoding="latin-1") as f:
    for line in f:
        line = line.rstrip()
        
        # Fixed: pass a tuple to startswith
        if line.startswith(("V", "P", "M", "S")):
            parts = line.split(maxsplit=1)
            var_name = parts[0]
            
            if len(parts) > 1:
                rest = parts[1]
                # Capture everything until just before metadata like "1 N", "2 A", etc.
                match = re.match(r"^(.*?)(?=\s+\d+\s+[A-Z]\s)", rest)
                if match:
                    var_label = match.group(1).strip()
                else:
                    var_label = rest.strip()
            else:
                var_label = var_name
            
            var_labels[var_name] = var_label

# Preview
for k, v in list(var_labels.items())[:15]:
    print(f"{k} → {v}")

# Total count
print(f"Total variables mapped: {len(var_labels)}")
```

```python
import pandas as pd

data = pd.read_csv("pregnancy_postnatal_care_numeric.csv")

# Rename columns using cleaned labels
data.rename(columns={col: var_labels.get(col.upper(), col) for col in data.columns}, inplace=True)

data.to_csv("data_renamed_numeric.csv", index=False)
print("✅ Clean column names applied!")
```

```python
import pandas as pd
import re

# Load your CSV file
data = pd.read_csv("data_renamed_numeric.csv")

# Clean column names: remove trailing spaces and numbers/metadata
def clean_col(col):
    # Remove trailing spaces and any digits/metadata at the end
    col = re.sub(r"\s+\d.*$", "", col).strip()
    return col

# Apply to all columns
data.columns = [clean_col(col) for col in data.columns]

# Save cleaned CSV
data.to_csv("data_renamed_numeric.csv", index=False)

print("✅ Column names cleaned and saved as data_cleaned.csv")
print(data.columns.tolist())
```
