```python
import pandas as pd
from IPython.display import display, Markdown

def show_all_sheets(input_file, skiprows=9):
    # Load the Excel file
    xls = pd.ExcelFile(input_file)
    sheet_names = xls.sheet_names

    # Iterate over each sheet and display its data, skipping descriptive rows
    for sheet_name in sheet_names:
        df = pd.read_excel(xls, sheet_name=sheet_name, skiprows=skiprows)
        display(Markdown(f"### Sheet: {sheet_name}"))
        display(df)
        display(Markdown("\n---\n"))

if __name__ == "__main__":
    input_file = "VAW Cambodia 2015 original full STATA9.xlsx"
    show_all_sheets(input_file)
```

```python
import pandas as pd
from openpyxl import Workbook

def split_excel_sheets(input_file):
    # Read all sheets from the input Excel file
    xls = pd.ExcelFile(input_file)
    sheet_names = xls.sheet_names  # List of all sheet names

    # We expect 17 sheets
    total_sheets = len(sheet_names)
    if total_sheets != 17:
        print(f"Warning: Expected 17 sheets but found {total_sheets}")

    # For each sheet, create a new Excel file with 17 sheets named sheet1 to sheet17
    for i, sheet_name in enumerate(sheet_names):
        # Create a new workbook
        wb = Workbook()
        # Remove the default sheet created by openpyxl
        default_sheet = wb.active
        if default_sheet is not None:
            wb.remove(default_sheet)

        # Add 17 sheets named sheet1 to sheet17
        for j in range(1, 18):
            wb.create_sheet(title=f"sheet{j}")

        # Load the data from the current sheet into a DataFrame
        df = pd.read_excel(xls, sheet_name=sheet_name)

        # Write the data to the corresponding sheet in the new workbook
        # Using openpyxl's worksheet
        ws = wb[f"sheet{i+1}"]

        # Write DataFrame to worksheet
        for r_idx, row in enumerate(df.itertuples(index=False), 1):
            for c_idx, value in enumerate(row, 1):
                ws.cell(row=r_idx+1, column=c_idx, value=value)

        # Write the header row
        for c_idx, col_name in enumerate(df.columns, 1):
            ws.cell(row=1, column=c_idx, value=col_name)

        # Save the new workbook
        output_file = f"sheet{i+1}.xlsx"
        wb.save(output_file)
        print(f"Saved {output_file}")

if __name__ == "__main__":
    input_file = "result yearbook2023.xlsx"
    split_excel_sheets(input_file)
```

```python
rice_types = ['Non-aromatic paddy', 'Aromatic paddy', 'Sticky paddy']
rice_df = df[df['S3_CROP__id'].isin(rice_types)]
rice_df
```

```python
import matplotlib.pyplot as plt

# Count the occurrences of each rice type
rice_counts = rice_df['S3_CROP__id'].value_counts()

# Create bar plot
plt.figure(figsize=(10, 6))
bars = plt.bar(rice_counts.index, rice_counts.values)

# Customize the plot
plt.title('Distribution of Rice Types')
plt.xlabel('Rice Type')
plt.ylabel('Count')
plt.xticks(rotation=45)

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height):,}',
             ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
